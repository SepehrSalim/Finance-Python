```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Functional Style
plt.plot(x, y)
#plt.show()  Just for using outside of Jupyter Notebook
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Functional Style')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```

```python
# Object Oriented Style
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('OO Style')
```

```python
fig = plt.figure() # OO Style
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.2,.5,.4,.3])
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # Creates a figure & its subplots
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'g')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout() # Prevents subplots overlapping
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(3, 2))
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'g')
plt.tight_layout()
```

```python
fig.savefig('my_figure.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, 'r', label='X Squared')
ax.plot(x, x**3, 'g', label='X Cubed')
ax.legend()
```
